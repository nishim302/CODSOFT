```python
import pandas as pd
import numpy as np
```

```python
c_df = pd.read_csv('creditcard.csv')
```

```python
c_df.sample(5)
```

```python
c_df['Class'].unique()
```

```python
c_df['Class'].value_counts()
```

```python
c_df.groupby('Class').mean()
```

```python
legal = c_df[c_df['Class'] == 0]
fraud = c_df[c_df['Class'] == 1]
print(legal.shape, fraud.shape)
```

```python
legal_sample = legal.sample(n=492)
legal_sample.shape
```

```python
legal_fraud_df = pd.concat([legal_sample, fraud], axis=0)
```

```python
legal_fraud_df.shape
```

```python
legal_fraud_df.sample(5)
```

```python
legal_fraud_df.describe()
```

```python
legal_fraud_df['Class'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(legal_fraud_df.drop(columns='Class'))
scaled_legal_fraud_df = pd.DataFrame(scaled_data, columns=legal_fraud_df.columns[:-1])
```

```python
scaled_legal_fraud_df.sample(5)
```

```python
X = scaled_legal_fraud_df
y = legal_fraud_df['Class']
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=2, test_size=0.2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
my_model = LogisticRegression()
my_model.fit(X_train, y_train)
```

```python
train_preds = my_model.predict(X_train)
test_preds = my_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Training Data Accuracy: ",accuracy_score(train_preds, y_train))
```

```python
print("Test Data Accuracy: ", accuracy_score(test_preds, y_test))
```
